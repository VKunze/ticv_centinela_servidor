```python
import pandas as pd

#datos_nivel = pd.read_csv('datos_nivel_test.csv', sep= ';')
#datos_lluvia = pd.read_csv('datos_lluvia_test.csv', sep= ';')
datos = pd.read_csv('Datos completos/Data v1.csv', sep= ';')  # completos/Data v1 croppedDatos test v1.csv

datos.head()
```

```python
# y estos son los resultados que se obtienen, en el mismo orden
target_data = datos.iloc['Nivel']

# cargamos las 4 combinaciones de las compuertas XOR
training_data = datos.iloc[:, 2:12]
#.drop(columns="t").drop(columns="Fecha")

training_data.head()
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense

def base_model():
    model = Sequential()
    model.add(Dense(16, input_dim=len(training_data.columns.tolist()), activation='tanh'))
    model.add(Dense(1, activation='tanh'))

    model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['accuracy'])
    return model
```

```python
# Get feature importance

from tensorflow.keras.wrappers.scikit_learn import KerasRegressor
import eli5
from eli5.sklearn import PermutationImportance

sci_model = KerasRegressor(build_fn=base_model, nb_epoch=100, batch_size=2, verbose=0)  
sci_model.fit(training_data, target_data)

perm = PermutationImportance(sci_model, random_state=1).fit(training_data, target_data)
eli5.show_weights(perm, feature_names = training_data.columns.tolist())   
```

```python
model=base_model()

model.fit(training_data, target_data, nb_epoch=150)

# evaluamos el modelo
scores = model.evaluate(training_data, target_data)

print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
#print (model.predict(training_data).round())
```
